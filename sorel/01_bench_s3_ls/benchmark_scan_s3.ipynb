````python
# Uses the ```sorel``` environment
import boto3
````

````python
BUCKET = 'sorel-20m'
PREFIX = '09-DEC-2020/binaries'
````

````python
s3 = boto3.resource('s3')

bucket = s3.Bucket(BUCKET)
````

````python
size = sum(1 for _ in bucket.objects.all())
````

````python
filtered_objects = bucket.objects.filter(Prefix=f'{PREFIX}/fff')
print(f'Found {len([o for o in filtered_objects])} objects')
#print('End')
````

````python
for o in filtered_objects:
    print(o.key)
````

````python
filtered_objects = bucket.objects.filter(Prefix=f'{PREFIX}/0')
for i, o in enumerate(filtered_objects):
    print(o.key)
    if i > 10:
        break
````
