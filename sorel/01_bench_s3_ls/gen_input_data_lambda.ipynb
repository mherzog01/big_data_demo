# Generate data by making many calls to a Lambda function

```python
import boto3
import datetime
```

```python
TARGET_BUCKET = 'sorel-20m-demo'
TARGET_PREFIX = 'input'
```

```python
def logmsg(msg):
    print(msg)

def logmsg01(msg):
    logmsg(f'{datetime.datetime.now():%Y-%m-%d %H:%M:%S}:  {msg}')

logmsg01('Process begins')    
```

```python
boto3.setup_default_session(profile_name='personal', region_name='us-east-1')
```

```python
for i in range(256):
    #https://stackoverflow.com/questions/12638408/decorating-hex-function-to-pad-zeros
    padding = 2
    print(i, hex(i), f'{i:0{padding}x}')
```

Invoke Lambda

```python
lambda_client = boto3.client('lambda')
```

```python
def invoke_gen_data(filter_chars):
    payload = '{"filter_chars":"' + filter_chars + '"}'
    response = lambda_client.invoke(
        FunctionName='sorel_gen_keys',  #arn:aws:lambda:us-east-1:883375387566:function:sorel_gen_keys
        InvocationType='Event',
        LogType='Tail',
        #ClientContext='string',
        Payload=bytes(payload, 'utf-8')
        #Qualifier='string'
    )
    return response
response = invoke_gen_data('0002')
response
```

```python
response['ResponseMetadata']['HTTPStatusCode'] != 202 or response['StatusCode'] != 202
```

```python
# TODO Monitor progress
response['Payload'].read().decode('utf-8')
```

```python
logmsg01('Starting jobs')
padding = 2
for i in range(256):
    filter_chars = f'{i:0{padding}x}'
    print(i, filter_chars)
    response = invoke_gen_data(filter_chars)
    if response['ResponseMetadata']['HTTPStatusCode'] != 202 or response['StatusCode'] != 202:
        raise ValueError
    # if i > 4:
    #     break
```

```python
s3 = boto3.resource('s3')
objects = s3.Bucket(TARGET_BUCKET).objects.filter(Prefix=TARGET_PREFIX)
object_list = [o for o in objects]
logmsg01(f'# objects = {len(object_list)}')
```
